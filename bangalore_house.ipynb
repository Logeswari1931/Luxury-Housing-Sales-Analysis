```python
tmp_df = house_df.copy()
tmp_df
```

```python
tmp_df['Micro_Market']=tmp_df['Micro_Market'].str.title().str.strip()
tmp_df['Configuration']=tmp_df['Configuration'].str.upper().str.strip()
tmp_df['NRI_Buyer'] = tmp_df['NRI_Buyer'].str.capitalize().str.strip()
tmp_df['Project_Name'] = tmp_df['Project_Name'].str.replace('_',' ',regex=False).str.strip()

tmp_df['Unit_Size_Sqft'] = tmp_df['Unit_Size_Sqft'].replace(-1, pd.NA)
tmp_df['Unit_Size_Sqft'] = pd.to_numeric(tmp_df['Unit_Size_Sqft'], errors='coerce')
tmp_df['Unit_Size_Sqft'] = tmp_df['Unit_Size_Sqft'].astype(float)

tmp_df['Purchase_Quarter'] = pd.to_datetime(tmp_df['Purchase_Quarter'], dayfirst=True, errors='coerce')

tmp_df['Ticket_Price_Cr'] = (tmp_df['Ticket_Price_Cr'].astype(str).str.replace('₹', '', regex=False).str.replace('Cr', '', regex=False).str.strip())
tmp_df['Ticket_Price_Cr'] = pd.to_numeric(tmp_df['Ticket_Price_Cr'], errors='coerce')

tmp_df
```

```python
tmp_df.info()
```

```python
tmp_df.describe()
```

```python
tmp_df.loc[tmp_df['Ticket_Price_Cr'] < 0, 'Ticket_Price_Cr'] = np.nan
```

```python
#print(tmp_df.info())
print(tmp_df.isnull().sum())
#tmp_df = tmp_df.dropna(subset=['Ticket_Price_Cr'])
```

```python
# Treating Null Values
tmp_df['Ticket_Price_Cr'] = tmp_df['Ticket_Price_Cr'].fillna(tmp_df['Ticket_Price_Cr'].median())
tmp_df['Unit_Size_Sqft'] = tmp_df['Unit_Size_Sqft'].fillna(tmp_df['Unit_Size_Sqft'].median())
tmp_df['Amenity_Score'] = tmp_df['Amenity_Score'].fillna(tmp_df['Amenity_Score'].median())
tmp_df['Buyer_Comments'] = tmp_df['Buyer_Comments'].fillna("No Comment")
```

```python
cat_cols = ['Builder', 'Micro_Market', 'Configuration',
            'Possession_Status', 'Sales_Channel', 'Buyer_Type']
for col in cat_cols:
    if col in tmp_df.columns:
        tmp_df[col] = tmp_df[col].fillna(tmp_df[col].mode()[0])
```

```python
#Price_per_Sqft
tmp_df['Price_per_Sqft'] = (tmp_df['Ticket_Price_Cr'] * 1e7) / tmp_df['Unit_Size_Sqft']

#Quarter_Number
tmp_df['Quarter_Number'] = tmp_df['Purchase_Quarter'].dt.quarter

#Booking_Flag
tmp_df['Booking_Flag'] = tmp_df['Transaction_Type'].apply(
    lambda x: 1 if str(x).lower() in ['primary', 'secondary'] else 0
)
tmp_df['NRI_Buyer'] = tmp_df['NRI_Buyer'].map({'Yes': 1, 'No': 0})

```
